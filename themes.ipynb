```python
# Create a dictionary with all parent themes as keys, and a list of all their sub-themes as values. 
# Here, you should only discern between a parent theme and any subtheme. 
# Thus, theme 50 would be a parent theme, and both theme 80 and 67 should be listed on the same level.
```

```python
import csv

path = "themes.csv"
with open(path,'r') as file:
    csv_reader = csv.reader(file)
    header = next(csv_reader)
    pc_dict ={}
    root_ids={}                #root_ids{theme_id : {set of child ids}}
    node_ids={}                #node_ids{theme_id : parent_id}
    c=0
    
    for row in csv_reader:     #id,name,parent_id
        theme_id = row[0]
        name = row[1]
        parent_id = row[2].strip()
        c += 1

        #divide ids into two directories - root_ids and node_ids
        if parent_id == "":
            root_ids[theme_id]=root_ids.get(theme_id,{''})
        else:
            node_ids[theme_id]=node_ids.get(theme_id,parent_id)
            
        
    file.seek(1)
    csv_reader = csv.reader(file)
    header = next(csv_reader)
    for row in csv_reader:
        theme_id = row[0]
        name = row[1]
        parent_id = row[2].strip()
        
        #print(theme_id,name,parent_id )
        if parent_id != "":   # is not blank
            #print("within if bock")
            set_id = theme_id
            parent = parent_id
            child_themes = []
            
            child_themes.append(set_id)
            while parent not in root_ids.keys():
                #print("within while bock")
                
                child_themes.append(set_id)
                set_id = parent
                parent = node_ids[set_id]
                
            root_ids[parent].update(child_themes)# attach new list of childs to the root_ids value set 
            #print(parent,root_ids[parent] )

    for i in root_ids:
        #root_ids[i].remove('')
        print(i, root_ids[i])
```
